```python
import cv2
import os
```

```python
faceCascade=cv2.CascadeClassifier("C:\\Users\\manoj\\OneDrive\\Desktop\\New folder\\haarcascade_frontalface_default.xml")
```

```python
image=cv2.imread(r"C:\Users\manoj\OneDrive\Desktop\New folder\boys.jpg")
```

```python
from matplotlib import pyplot as plt 
plt.imshow(image)
```

```python
RGB_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
GR_img=cv2.cvtColor(RGB_img,cv2.COLOR_BGR2GRAY)
plt.imshow(GR_img)
```

```python
faces=faceCascade.detectMultiScale(
    GR_img,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30,30)
)

print(faces)
```

```python
image=cv2.imread(r"C:\Users\manoj\OneDrive\Desktop\New folder\boys.jpg")
for (x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),10)
RGB_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```
